**Exercise:** The numbers generated by `numpy.random.random` are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from `numpy.random.random` and plot their PMF.  What goes wrong?

Now plot the CDF. Is the distribution uniform?

```python
# Solution is:
t = np.random.random(1000)
```

```python
# Solution is:
pmf = thinkstats2.Pmf(t)
thinkplot.Pmf(pmf, linewidth=0.1)
thinkplot.Config(xlabel='Random variate', ylabel='PMF')/n/nBased on the PMF graph generated, the final outcome is straight and smooth, since there are not much spikes and valleys. So no much differences between the distributions.
```

```python
# Solution is:
cdf = thinkstats2.Cdf(t)
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Random variate', ylabel='CDF')/n/n Based on the CDF report generated, the final outcome is distributed uniform. The CDF is approximately a straight line, which means that the distribution is uniform.
```
